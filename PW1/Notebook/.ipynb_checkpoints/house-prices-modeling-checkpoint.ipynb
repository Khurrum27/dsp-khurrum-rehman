```python
from sklearn.metrics import mean_squared_log_error
import numpy as np

def compute_rmsle(y_test, y_pred, precision=4):
    # Ensure no NaN or infinite values in y_pred and y_test
    y_pred = np.nan_to_num(y_pred, nan=np.nanmean(y_pred))  # Replace NaNs in predictions with the mean value
    y_test = np.nan_to_num(y_test, nan=np.nanmean(y_test))  # Replace NaNs in actual values with the mean value
    
    # Clip negative or zero values to a small positive number
    y_test = np.clip(y_test, a_min=1e-5, a_max=None)
    y_pred = np.clip(y_pred, a_min=1e-5, a_max=None)

    # Apply log1p safely
    y_test_log = np.log1p(y_test)  # log(1 + y_test)
    y_pred_log = np.log1p(y_pred)  # log(1 + y_pred)
    
    # Compute RMSLE
    rmsle = np.sqrt(mean_squared_log_error(y_test_log, y_pred_log))
    return round(rmsle, precision)

# Example usage (replace y_test and y_pred with your actual values):
rmsle_score = compute_rmsle(y_test, y_pred)
print(f'RMSLE Score: {rmsle_score}')
```
